```python
#Q: Find the highest positive correlation among all attributes. While finding the correlation, use appropriate
#code, not manually. Print this information accordingly.

def lab04_task10_2019_1_60_120():

    Correlation_positive= df.corr() 
    cor_pos= Correlation_positive[Correlation_positive>0]
    print("All attributes are : ",cor_pos.index)
    print("Here the highest positive correlation among all attributes is : ",cor_pos.index.max())
    print("\n")

    #print(cor_pos)
    plt.figure(figsize= (30,15))

    sns.heatmap(cor_pos, cmap="Greens",annot=True)

lab04_task10_2019_1_60_120()
```
